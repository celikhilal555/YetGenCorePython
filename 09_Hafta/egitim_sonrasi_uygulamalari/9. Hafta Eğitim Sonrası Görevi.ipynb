# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız

import random
import time

class Kumanda:
    def __init__(self):
        self.tv_durumu = False
        self.ses_seviyesi = 50
        self.kanal_listesi = []
        self.aktif_kanal = None

    def tv_ac(self):
        if not self.tv_durumu:
            print("Televizyon açılıyor...")
            time.sleep(1)
            self.tv_durumu = True
            print("Televizyon açıldı.")
        else:
            print("Televizyon zaten açık.")

    def tv_kapat(self):
        if self.tv_durumu:
            print("Televizyon kapatılıyor...")
            time.sleep(1)
            self.tv_durumu = False
            print("Televizyon kapandı.")
        else:
            print("Televizyon zaten kapalı.")

    def ses_ayarla(self, seviye):
        if 0 <= seviye <= 100:
            self.ses_seviyesi = seviye
            print(f"Ses seviyesi {self.ses_seviyesi} olarak ayarlandı.")
        else:
            print("Geçersiz ses seviyesi. Ses seviyesi 0-100 arasında olmalıdır.")

    def kanal_ekle(self, kanal):
        if kanal not in self.kanal_listesi:
            self.kanal_listesi.append(kanal)
            print(f"{kanal} kanalı eklendi.")
        else:
            print(f"{kanal} kanalı zaten ekli.")

    def kanal_sayisini_ogren(self):
        kanal_sayisi = len(self.kanal_listesi)
        print(f"Toplam {kanal_sayisi} kanal bulunuyor.")

    def rastgele_kanala_gec(self):
        if self.kanal_listesi:
            self.aktif_kanal = random.choice(self.kanal_listesi)
            print(f"Rastgele bir kanala geçildi: {self.aktif_kanal}")
        else:
            print("Kanal bulunamadı. Lütfen önce kanal ekleyin.")

    def televizyon_bilgileri(self):
        print("----- Televizyon Bilgileri -----")
        print(f"TV Durumu: {'Açık' if self.tv_durumu else 'Kapalı'}")
        print(f"Ses Seviyesi: {self.ses_seviyesi}")
        print(f"Aktif Kanal: {self.aktif_kanal}")
        print(f"Kanal Listesi: {self.kanal_listesi}")
        print("---------------------------------")

    def baslat(self):
        while True:
            print("----- KUMANDA -----")
            print("1. TV Aç")
            print("2. TV Kapat")
            print("3. Ses Ayarları")
            print("4. Kanal Ekle")
            print("5. Kanal Sayısını Öğrenme")
            print("6. Rastgele Kanala Geçme")
            print("7. Televizyon Bilgileri")
            print("q. Çıkış")
            secim = input("Yapmak istediğiniz işlemi seçin: ")

            if secim == "1":
                self.tv_ac()
            elif secim == "2":
                self.tv_kapat()
            elif secim == "3":
                yeni_ses_seviyeleri = int(input("Yeni ses seviyesini girin (0-100): "))
                self.ses_ayarla(yeni_ses_seviyeleri)
            elif secim == "4":
                yeni_kanal = input("Eklemek istediğiniz kanalı girin: ")
                self.kanal_ekle(yeni_kanal)
            elif secim == "5":
                self.kanal_sayisini_ogren()
            elif secim == "6":
                self.rastgele_kanala_gec()
            elif secim == "7":
                self.televizyon_bilgileri()
            elif secim == "q":
                print("Kumanda kapatılıyor...")
                time.sleep(1)
                print("Kumanda kapatıldı.")
                break
            else:
                print("Geçersiz bir seçenek girdiniz. Lütfen tekrar deneyin.")
    
kumanda=Kumanda()
kumanda.baslat()
```
