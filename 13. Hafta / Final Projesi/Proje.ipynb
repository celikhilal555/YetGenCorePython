```python

# Karbon ayak izi hakkında kullanıcıyı bilgilendirme mesajı
print("Merhaba! Sizlere karbon ayak izi hakkında bilgi vermek istiyoruz. Karbon ayak izi, bir kişinin veya bir organizasyonun sera gazı emisyonlarının miktarını ölçen bir metriktir.")

# Karbon ayak izi hesaplamak için kullanılacak formülleri kullanıcıya sunma
print("Karbon ayak izinizi hesaplamak için çeşitli formüller vardır, ancak genellikle kişisel enerji tüketimi, taşımacılık, yiyecek tüketimi ve diğer faktörler göz önünde bulundurulur.")

# Ürünlerin listesini dosyaya yazma
with open("karbon_ayakizi.txt", "a") as f:
    f.write("Karbon Ayakizine Etki Eden Faktörler:\n")
    f.write("- Ulaşım\n")
    f.write("- Gıda\n")
    f.write("- Su\n")
    f.write("- Ev\n")

# Kullanıcıya mesaj gösterme
print("Listedeki bu ürünler, karbon ayak izinizi hesaplamada önemli faktörlerdir. Bunlardan herhangi birini kullanırken daha sürdürülebilir seçenekleri tercih ederek karbon ayak izinizi azaltabilirsiniz. İyi günler dileriz!\n\n")

        

class CarbonFootprintTransport:
    def __init__(self, km_yil, arac_turu, arac_yasi, yakit_turu, elektrik_fatura):
        self.km_yil = km_yil
        self.arac_turu = arac_turu
        self.arac_yasi = arac_yasi
        self.yakit_turu = yakit_turu
        self.elektrik_fatura = elektrik_fatura
    
    def calculate_transport_carbon_footprint(self):
        if self.arac_turu == "otomobil":
            if self.arac_yasi < 5:
                karbon_salim = 0.15 * self.km_yil
            elif self.arac_yasi >= 5 and self.arac_yasi < 10:
                karbon_salim = 0.20 * self.km_yil
            else:
                karbon_salim = 0.25 * self.km_yil
        else:
            if self.arac_yasi < 5:
                karbon_salim = 0.10 * self.km_yil
            elif self.arac_yasi >= 5 and self.arac_yasi < 10:
                karbon_salim = 0.15 * self.km_yil
            else:
                karbon_salim = 0.20 * self.km_yil
        
        if self.yakit_turu == "benzin":
            karbon_salim = karbon_salim * 2.32
        elif self.yakit_turu == "dizel":
            karbon_salim = karbon_salim * 2.63
        else:
            karbon_salim = karbon_salim * 2.77
            
        return karbon_salim
    
    def calculate_electricity_carbon_footprint(self):
        elektrik_kullanimi = self.elektrik_fatura / 0.15
        elektrik_karbon = elektrik_kullanimi * 0.527
        return elektrik_karbon
    
    def calculate_total_carbon_footprint(self):
        transport_carbon_footprint = self.calculate_transport_carbon_footprint()
        electricity_carbon_footprint = self.calculate_electricity_carbon_footprint()
        total_carbon_footprint = transport_carbon_footprint + electricity_carbon_footprint
        return total_carbon_footprint

    


class CarbonFootprintFood:
    def __init__(self, yillik_sebze, yillik_meyve, yillik_et, yillik_sut, yillik_diger):
        self.yillik_sebze = yillik_sebze
        self.yillik_meyve = yillik_meyve
        self.yillik_et = yillik_et
        self.yillik_sut = yillik_sut
        self.yillik_diger = yillik_diger

    def calculate_vegetable_carbon_footprint(self):
        vegetable_emission = self.yillik_sebze * 0.06 # kg CO2 eşdeğeri/kg
        return vegetable_emission

    def calculate_fruit_carbon_footprint(self):
        fruit_emission = self.yillik_meyve * 0.32 # kg CO2 eşdeğeri/kg
        return fruit_emission

    def calculate_meat_carbon_footprint(self):
        meat_emission = self.yillik_et * 7.6 # kg CO2 eşdeğeri/kg
        return meat_emission

    def calculate_dairy_carbon_footprint(self):
        dairy_emission = self.yillik_sut * 1.9 # kg CO2 eşdeğeri/kg
        return dairy_emission

    def calculate_other_carbon_footprint(self):
        other_emission = self.yillik_diger * 1.4 # kg CO2 eşdeğeri/kg
        return other_emission

    def calculate_total_carbon_footprint(self):
        vegetable_carbon_footprint = self.calculate_vegetable_carbon_footprint()
        fruit_carbon_footprint = self.calculate_fruit_carbon_footprint()
        meat_carbon_footprint = self.calculate_meat_carbon_footprint()
        dairy_carbon_footprint = self.calculate_dairy_carbon_footprint()
        other_carbon_footprint = self.calculate_other_carbon_footprint()
        total_carbon_footprint = (vegetable_carbon_footprint + fruit_carbon_footprint 
                                 + meat_carbon_footprint + dairy_carbon_footprint 
                                 + other_carbon_footprint)
        return total_carbon_footprint



class CarbonFootprintWater:
    def __init__(self, yillik_su_tuketimi):
        self.yillik_su_tuketimi = yillik_su_tuketimi
        self.litre_basina_emisyon = 0.2
    def calculate_water_carbon_footprint(self):
         
        su_emisyonu = self.yillik_su_tuketimi * self.litre_basina_emisyon
        return su_emisyonu

    def calculate_total_carbon_footprint(self):
        su_carbon_footprint = self.calculate_water_carbon_footprint()
        return su_carbon_footprint


class EvKarbonAyakizi:
    def __init__(self, dogalgaz, elektrik):
        self.dogalgaz = dogalgaz
        self.elektrik = elektrik

    def dogalgaz_karbon_ayakizi(self, yil, metrekare, kisi_sayisi):
        m3_yil = self.dogalgaz
        kg_c02 = 0.21 * m3_yil * yil * metrekare * kisi_sayisi
        return kg_c02

    def elektrik_karbon_ayakizi(self, yil, kisi_sayisi):
        kwh_yil = self.elektrik
        kg_c02 = 0.0006 * kwh_yil * yil * kisi_sayisi
        return kg_c02

    def ev_karbon_ayakizi(self, yil, metrekare, kisi_sayisi):
        dogalgaz_ayakizi = self.dogalgaz_karbon_ayakizi(yil, metrekare, kisi_sayisi)
        elektrik_ayakizi = self.elektrik_karbon_ayakizi(yil, kisi_sayisi)
        return dogalgaz_ayakizi + elektrik_ayakizi
        
def main():
    try:
        
        print("***********************************************")
        print("           ULAŞIM KARBON AYAK İZİ              ")
        print("***********************************************\n\n")
        
        km_yil = float(input("Yılda kaç kilometre yol alıyorsunuz? "))
        arac_turu = input("Araç türünüz (otomobil, motosiklet, vb.): ")
        arac_yasi = int(input("Araç yaşınız nedir? "))
        yakit_turu = input("Kullandığınız yakıt türü (benzin, dizel, vb.): ")
        elektrik_fatura = float(input("Aylık elektrik faturanız nedir? "))
        
        carbon_footprint = CarbonFootprintTransport(km_yil, arac_turu, arac_yasi, yakit_turu, elektrik_fatura)
        total_carbon_footprint_transport = carbon_footprint.calculate_total_carbon_footprint()
        print("Yıllık ulaşım karbon ayak iziniz: ", total_carbon_footprint_transport, "kg CO2 eşdeğeri\n\n")
        print("***********************************************")
        print("             GIDA KARBON AYAK İZİ              ")
        print("***********************************************\n\n")
        
        
        yillik_sebze = float(input("Yılda kaç kilogram sebze tüketiyorsunuz? "))
        yillik_meyve = float(input("Yılda kaç kilogram meyve tüketiyorsunuz? "))
        yillik_et = float(input("Yılda kaç kilogram et tüketiyorsunuz? "))
        yillik_sut = float(input("Yılda kaç kilogram süt tüketiyorsunuz? "))
        yillik_diger = float(input("Yılda kaç kilogram diğer gıda türleri tüketiyorsunuz? "))

        carbon_footprint = CarbonFootprintFood(yillik_sebze, yillik_meyve, yillik_et, yillik_sut, yillik_diger)
        total_carbon_footprint_food= carbon_footprint.calculate_total_carbon_footprint()

        print(f"Yıllık sebze karbon ayak izi: {carbon_footprint.calculate_vegetable_carbon_footprint():.2f} kg CO2 eşdeğeri")
        print(f"Yıllık meyve karbon ayak izi: {carbon_footprint.calculate_fruit_carbon_footprint():.2f} kg CO2 eşdeğeri")
        print(f"Yıllık et karbon ayak izi: {carbon_footprint.calculate_meat_carbon_footprint():.2f} kg CO2 eşdeğeri")
        print(f"Yıllık süt karbon ayak izi: {carbon_footprint.calculate_dairy_carbon_footprint():.2f} kg CO2 eşdeğeri")
        print(f"Yıllık diğer gıdaların karbon ayak izi: {carbon_footprint.calculate_other_carbon_footprint():.2f} kg CO2 eşdeğeri")
        print(f"Yıllık gıda karbon ayak iziniz: {total_carbon_footprint_food:.2f} kg CO2 eşdeğeri\n\n")
        print("***********************************************")
        print("              SU KARBON AYAK İZİ               ")
        print("***********************************************\n\n")

    
        yillik_su_tuketimi = float(input("Yılda kaç litre su tüketiyorsunuz? "))

        carbon_footprint = CarbonFootprintWater(yillik_su_tuketimi)
        su_carbon_footprint = carbon_footprint.calculate_water_carbon_footprint()
        toplam_carbon_footprint_water = carbon_footprint.calculate_total_carbon_footprint()

        print(f"Yıllık su tüketiminiz: {yillik_su_tuketimi:.2f} litre")
        print(f"Litre başına karbon ayak izi: {carbon_footprint.litre_basina_emisyon:.2f} g CO2 eşdeğeri")
        print(f"Toplam su karbon ayak izi: {su_carbon_footprint:.2f} g CO2 eşdeğeri")
        print(f"Toplam karbon ayak iziniz: {toplam_carbon_footprint_water:.2f} g CO2 eşdeğeri\n\n")
        print("***********************************************")
        print("             EV KARBON AYAK İZİ                ")
        print("***********************************************\n\n")

        
        dogalgaz = float(input("Yılda kaç metreküp doğalgaz tüketiyorsunuz? "))
        elektrik = float(input("Yılda kaç kWh elektrik tüketiyorsunuz? "))
        metrekare = float(input("Ev kaç metrekare? "))
        kisi_sayisi = int(input("Evde kaç kişi yaşıyor? "))
        yil = int(input("Ev kaç yıllık ? "))

        carbon_footprint = EvKarbonAyakizi(dogalgaz,elektrik)
        ev_karbon_ayakizi = carbon_footprint.ev_karbon_ayakizi(yil,metrekare,kisi_sayisi)

        print(f"Yıllık doğalgaz karbon ayak izi: {carbon_footprint.dogalgaz_karbon_ayakizi(yil,metrekare, kisi_sayisi):.2f} kg CO2 eşdeğeri")
        print(f"Yıllık elektrik karbon ayak izi: {carbon_footprint.elektrik_karbon_ayakizi(yil,kisi_sayisi):.2f} kg CO2 eşdeğeri")
        print(f"Yıllık ev karbon ayak iziniz: {ev_karbon_ayakizi:.2f} kg CO2 eşdeğeri\n\n")
        
        print("***********************************************")
        print("            TOPLAM KARBON AYAK İZİ             ")
        print("***********************************************\n\n")
        
        print("Toplam Karbon Ayak İziniz: " ,toplam_carbon_footprint_water + total_carbon_footprint_food + total_carbon_footprint_transport+ ev_karbon_ayakizi)
        
        
    except ValueError:
        print("Lütfen sadece sayısal değer girin.")

main()
```
