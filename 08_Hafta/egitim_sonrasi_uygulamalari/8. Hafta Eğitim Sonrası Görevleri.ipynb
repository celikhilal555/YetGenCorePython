# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız

num_of_entries = int(input("Kaç adet veri girişi yapacaksınız? "))
data = []

for i in range(num_of_entries):
    name_surname = input("{}. kullanıcının adı ve soyadı: ".format(i+1))
    age = int(input("{}. kullanıcının yaşı: ".format(i+1)))
    height = float(input("{}. kullanıcının boyu (metre cinsinden): ".format(i+1)))
    weight = float(input("{}. kullanıcının kilosu (kg cinsinden): ".format(i+1)))
    
    # VKI hesaplama
    bmi = weight / (height**2)
    
    # VKI'ya göre sınıflandırma
    if bmi < 18.5:
        category = "Zayıf"
    elif 18.5 <= bmi < 25:
        category = "Normal kilolu"
    elif 25 <= bmi < 30:
        category = "Fazla kilolu"
    elif 30 <= bmi < 40:
        category = "Obez"
    else:
        category = "İleri derecede obez"
    
    # verileri kaydetme
    entry = {"Ad Soyad": name_surname, "Yaş": age, "Boy": height, "Kilo": weight, "VKI": bmi, "Sınıf": category}
    data.append(entry)

import csv

# dosya adı
file_name = "veriler.csv"

# alan adları
fields = ["Ad Soyad", "Yaş", "Boy", "Kilo", "VKI", "Sınıf"]

# dosyayı açma ve verileri yazma
with open(file_name, 'w', newline='') as csv_file:
    writer = csv.DictWriter(csv_file, fieldnames=fields)
    writer.writeheader()
    for entry in data:
        writer.writerow(entry)
        
print("Veriler başarıyla kaydedildi.")
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
# 2.Problemin Çözümünü Buraya Yazınız

import datetime

def menu():
    print("YetBar Menüsü")
    print("1. Menüyü Görüntüle")
    print("2. Satın Al")
    print("3. Satın Alma Geçmişi")
    print("4. Çıkış")

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

class Transaction:
    def __init__(self, product, price):
        self.product = product
        self.price = price
        self.date = datetime.datetime.now()

    def __str__(self):
        return f"{self.product.name} - {self.price} TL - {self.date.strftime('%d/%m/%Y %H:%M:%S')}"

class YetBar:
    def __init__(self):
        self.products = [
            Product("Enerji Bombası", 12),
            Product("Detoks Karışımı", 15),
            Product("Yeşil İlham", 10),
            Product("Protein Gücü", 20),
            Product("Süper Meyve", 18)
        ]
        self.transactions = []

    def display_menu(self):
        menu()
    
    def display_products(self):
        for product in self.products:
            print(f"{product.name} - {product.price} TL")

    def buy(self):
        selected_product_name = input("Ürün seçiniz: ")
        selected_product = next((p for p in self.products if p.name == selected_product_name), None)
        if selected_product:
            print(f"{selected_product.name} - {selected_product.price} TL")
            transaction = Transaction(selected_product, selected_product.price)
            self.transactions.append(transaction)
            print("Satın alma işlemi başarıyla tamamlandı.")
        else:
            print("Ürün mevcut değil.")

    def display_transactions(self):
        for transaction in self.transactions:
            print(transaction)

    def run(self):
        while True:
            self.display_menu()
            choice = input("Seçiminiz: ")
            if choice == "1":
                self.display_products()
            elif choice == "2":
                self.buy()
            elif choice == "3":
                self.display_transactions()
            elif choice == "4":
                break
            else:
                print("Geçersiz seçim.")

YetBar().run()



```
